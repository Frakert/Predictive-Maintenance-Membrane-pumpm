### devide

```python
def devide(x,y):
    if ((x == 0) | (y == 0)):
        return 0
    else: 
        return x/y
```

### Confuion

```python
def Confusion(y_real,y_pred,silent):
    from sklearn import metrics 
    from sklearn.metrics import confusion_matrix
    import matplotlib.pyplot as plt
    
    cf=confusion_matrix(y_real,y_pred)
    TN=cf[0][0]
    TP=cf[1][1]
    FP=cf[0][1]
    FN=cf[1][0]

    Recall=devide(TP,(TP+FN))
    Precision=devide(TP,(TP+FP))
    F1=2*devide((Recall*Precision),(Precision+Recall))
    
    
    
    if silent != 1:
        print(F1)
        cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cf, display_labels = [False, True])
        cm_display.plot()
        plt.show()

    return(F1)
```

### Train_fit_Validate_model

```python
def Train_fit_Validate_model(X_train,y_train,X_val,y_val,model,plot):
    model = model

    model.fit(X_train,y_train)


    y_fit=model.predict(X_train)
    y_pred=model.predict(X_val)

    if plot == 1:
        print('Test Results:')
        F1=Confusion(y_val,y_pred,silent=0)

        print(' ')
        print('Train results:')
        Confusion(y_train,y_fit,silent=0)
        
    elif plot == 0:
        F1=Confusion(y_val,y_pred,silent=1)

    return F1
```

### train_and_test_kfold

```python
def train_and_test_kfold(X_train_val,y_train_val,model,n_splits,plot):

    import numpy as np
    from sklearn.model_selection import StratifiedKFold
    from imblearn.over_sampling import SMOTE
    
    F1_array=[]

    kf = StratifiedKFold(n_splits=n_splits,shuffle=True)
    for train_index,test_index in kf.split(X_train_val,y_train_val):
        
        X_train, X_test = X_train_val.iloc[train_index], X_train_val.iloc[test_index]
        y_train, y_test = y_train_val.iloc[train_index], y_train_val.iloc[test_index]

        oversample = SMOTE()
        X_train_o, y_train_o = oversample.fit_resample(X_train, y_train)


        F1=Train_fit_Validate_model(X_train_o,y_train_o,X_val,y_val, model=model, plot=plot)
        F1_array.append(F1)
    
    return F1_array
```
